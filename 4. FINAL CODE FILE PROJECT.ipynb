```python
Tourist_Type = []
for index, row in df_merged.iterrows():
    if (row['CATEGORY'] == 'Art Gallery'):
        Value = 'Family, Couple, Individual'
        Tourist_Type.append(Value)
    elif (row['CATEGORY'] == 'Museums'):
        Value = 'Family, Couple, Individual'
        Tourist_Type.append(Value)
    elif (row['CATEGORY'] == "Botanical Garden"):
        Value = 'Family, Couple'
        Tourist_Type.append(Value)
    elif (row['CATEGORY'] == "Water Trails"):
        Value = 'Friends, Couple, Individual'
        Tourist_Type.append(Value)
    elif (row['CATEGORY'] == "Beaches"):
        Value = 'Family, Friends, Couple, Individual'
        Tourist_Type.append(Value)
    elif (row['CATEGORY'] == "Hiking Trail"):
        Value = 'Friends, Couple, Individual'
        Tourist_Type.append(Value)
df_merged["TOURIST TYPE"] = Tourist_Type
```

### Cleanup

#### Note: Here we did not handle nan values because the number of zipcodes near different attractions differ and hence the remaining number of rows are filled with NaN which we need. For the code to work well we replace it with 0.


```python
#change datatype of ZipCode column to int 
#Fill `NaN` values in Zip Code to 0 

columns = ['ZIP Code_Statue of Liberty', 'ZIP Code_Empire State Building', 'ZIP Code_Times Square', 'ZIP Code_Central Park']
df_zipcode['ZIP Code_Statue of Liberty'] = pd.to_numeric(df_zipcode['ZIP Code_Statue of Liberty'], errors='coerce')
df_zipcode['ZIP Code_Empire State Building'] = pd.to_numeric(df_zipcode['ZIP Code_Empire State Building'], errors='coerce')
df_zipcode['ZIP Code_Times Square'] = pd.to_numeric(df_zipcode['ZIP Code_Times Square'], errors='coerce')
df_zipcode['ZIP Code_Central Park'] = pd.to_numeric(df_zipcode['ZIP Code_Central Park'], errors='coerce')
df_zipcode[columns] = df_zipcode[columns].fillna(0)
```

```python
df_zipcode
```

### Function to filter the Merged datasets based on Major Attraction 

```python
#function to return the values based on tourist attraction for Merged dataset

def get_filtered_tourist_df_by_attraction_zipcode(attraction_name):
    zip_column_name = 'ZIP Code_' + attraction_name
    distance_column_name = 'Distance_' + attraction_name
    list_of_zipcode_close_to_attraction = df_zipcode[zip_column_name].to_list()
    mask = df_merged['ZIPCODE'].isin(list_of_zipcode_close_to_attraction)
    filtered_df_merged = df_merged[mask]
    df_result = pd.merge(filtered_df_merged, df_zipcode[[zip_column_name, distance_column_name]],  how='left', left_on=['ZIPCODE'], right_on = [zip_column_name])
    return df_result.drop(zip_column_name, axis=1)
```

### Function to filter the restaurant datasets based on attraction

```python
#function to return the closest restaurants for restaurant dataset

def get_filtered_restaurants(attraction_name):
    zip_col_nm = 'ZIP Code_' + attraction_name
    dist_col_nm = 'Distance_' + attraction_name
    list_of_zipcode_close_to_restaurant = df_zipcode[zip_col_nm].to_list()
    mask_val = df_restaurant['ZIPCODE'].isin(list_of_zipcode_close_to_restaurant)
    filtered_df_restaurant = df_restaurant[mask_val]
    df_result = pd.merge(filtered_df_restaurant, df_zipcode[[zip_col_nm, dist_col_nm]],  how='left', left_on=['ZIPCODE'], right_on = [zip_col_nm])
    return df_result.drop(zip_col_nm, axis=1)
```

### Function to filter the datasets based on distance

```python
#function to return the values based on distance for merged dataset
def get_filtered_attraction_df_by_distance(closeby_attractions_df, attraction_name, upper_limit_distance):
    distance_column_name = 'Distance_' + attraction_name
    closeby_attractions['distance_value'] = closeby_attractions_df.apply(lambda x: float(x[distance_column_name].split(' ')[0]), axis=1)
    distance_limited_attractions = closeby_attractions[closeby_attractions['distance_value'] <= upper_limit_distance]
    distance_limited_attractions.sort_values(by='distance_value', inplace=True)
    filterdist = distance_limited_attractions.drop('distance_value', axis=1)
    return filterdist
```

```python
#function to return the values based on distance for restaurant
def get_filtered_restaurant_by_distance(closeby_restaurants_df, attraction_name, upper_limit_distance):
    dist_col_name = 'Distance_' + attraction_name
    closeby_restaurants_df['distance_value'] = closeby_restaurants_df.apply(lambda x: float(x[dist_col_name].split(' ')[0]), axis=1)
    distance_limited_restaurants = closeby_restaurants_df[closeby_restaurants_df['distance_value'] <= upper_limit_distance]
    distance_limited_restaurants.sort_values(by='distance_value', inplace=True)
    filterdistrestaurant = distance_limited_restaurants.drop('distance_value', axis=1)
    return filterdistrestaurant
```

```python
#testing if the function works properly on call
closeby_attractions = get_filtered_tourist_df_by_attraction_zipcode('Central Park')
attraction_distance_output = get_filtered_attraction_df_by_distance(closeby_attractions, 'Central Park', 2)
```

### Filtering based on Tourist Type

```python
#function to filter the dataframes based on the tourist type based on the user input.

def tourist_type_filter(tourist_type):
    df_tt_updated = attraction_distance_output[attraction_distance_output['TOURIST TYPE'].str.contains(tourist_type)]
    return df_tt_updated
```

```python
import pandas as pd
import PySimpleGUI as psg
import io
import os
from PIL import Image
```

```python
l_majorattractions = ["Central Park", "Empire State Building", "Statue of Liberty", "Times Square"]
l_distance = [1, 2, 3, 4, 5]
l_touristtypes = ["Individual", "Couple", "Family", "Friends"]
```

```python
image1 = Image.open(r'New York City - 1.jpeg')
image1.thumbnail((200, 200))
bio1 = io.BytesIO()
image1.save(bio1, format="PNG")

image2 = Image.open(r'New York City - 2.jpeg')
image2.thumbnail((200, 200))
bio2 = io.BytesIO()
image2.save(bio2, format="PNG")

image3 = Image.open(r'New York City - 3.jpeg')
image3.thumbnail((200, 200))
bio3 = io.BytesIO()
image3.save(bio3, format="PNG")

image4 = Image.open(r'New York City - 4.jpeg')
image4.thumbnail((200, 200))
bio4 = io.BytesIO()
image4.save(bio4, format="PNG")
```

```python
layout = [[psg.Text("Tell us about your preferences and we will recommend the best places to visit to make your trip unforgettable!", size=(100, 2), font='Lucida', justification='left')],
          [psg.Image(data=bio1.getvalue(), key="-IMAGE-"),
          psg.Image(data=bio2.getvalue(), key="-IMAGE-"),
          psg.Image(data=bio3.getvalue(), key="-IMAGE-"),
          psg.Image(data=bio4.getvalue(), key="-IMAGE-")],
          [psg.Text("Major Attraction:", size=(60, 1), font='Lucida', justification='left')], 
          [psg.Combo(l_majorattractions)],
          [psg.Text("Distance (in miles):", size=(60, 1), font='Lucida', justification='left')], 
          [psg.Combo(l_distance)],
          [psg.Text("Tourist Type:", size=(50, 1), font='Lucida', justification='left')],
          [psg.Combo(l_touristtypes)],
          [psg.Button("RECOMMEND", font=("Lucida", 10), key="Recommend"), psg.Button("CANCEL", font=("Lucida", 10), key="Cancel")]
    ]
```

```python
win = psg.Window(title="Welcome!", layout=layout)
```

```python
action, inputs = win.read()
ma_var = inputs[0]
d_var = inputs[1]
tt_var = inputs[2]
print(ma_var, d_var, tt_var)

while True:
    if action == "Cancel":
        break
    elif action == "Recommend":
        #call functions based on input values
        closeby_attractions = get_filtered_tourist_df_by_attraction_zipcode(ma_var)
        attraction_distance_output = get_filtered_attraction_df_by_distance(closeby_attractions, ma_var, d_var)
        op = tourist_type_filter(tt_var)
        final_output = op.drop(columns=['TOURIST TYPE'])
        print(final_output)
        
        #call restaurnat filter function
        restaurant_bydist= get_filtered_restaurants(ma_var)
        restaurant_op = get_filtered_restaurant_by_distance(restaurant_bydist, ma_var, d_var)
        
        data = []
        headers = []
        #headers = final_output.iloc[0].tolist()
        headers = final_output.columns.values.tolist()
        
        data = final_output[1:].values.tolist()
        
        restaurant_data =[]
        headers_r =[]
        headers_r =  restaurant_op.columns.values.tolist()
        restaurant_data =  restaurant_op[1:].values.tolist()
        
        
        #new op window
        output_layout = [
        [psg.Text("Here are the places you can visit:", size=(100, 2), font='Lucida', justification='left')],
        [
            psg.Table(values=data, headings = headers, 
                      auto_size_columns= False, 
                      justification='right',
                      num_rows=10,
                     tooltip='Recommendations for NYC')
        ],
        [psg.Text("Hungry after travelling so much? You can eat here:", size=(100, 2), font='Lucida', justification='left')],
        [
            psg.Table(values=restaurant_data, headings=headers_r, 
                      auto_size_columns= False, 
                      justification='right',
                      num_rows=10,
                     tooltip='Recommendations for NYC')
        ],
        ]

        opwindow = psg.Window(title = 'Recommended Places to visit', layout = output_layout)

        while True:
            event, values = opwindow.read()
            if event == "Cancel" or event == sg.WIN_CLOSED:
                break

        opwindow.close()
       
```
