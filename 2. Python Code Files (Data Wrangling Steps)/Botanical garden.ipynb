```python
# Import the required libraries
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
#Reading the dataset

df_botanical_garden = pd.read_csv("/Users/ramyakumaresan/Downloads/Botanical Gardens.csv")
df_botanical_garden
```

```python
#Profiling the dataset

profile= ProfileReport(df_botanical_garden, title="profile")
profile.to_file("botanical_report.html")
```

```python
#Renaming the columns of the dataset

df_botanical_garden = df_botanical_garden.rename(columns={"NAME":"NAME","TEL":"PHONE NO.","ADDRESS1":"ADDRESS","ZIP":"ZIPCODE"})
df_botanical_garden
```

```python
#Adding a category column for "Botanical Garden"

df_botanical_garden['CATEGORY'] = "Botanical Garden"
df_botanical_garden
```

```python
#Dropping the columns that are not required in the dataset

df_botanical_garden.drop("the_geom",axis =1,inplace=True)
df_botanical_garden.drop("ADDRESS2",axis =1,inplace=True)
df_botanical_garden.drop("CITY",axis =1,inplace=True)
df_botanical_garden
```

```python
#Adding the additional details column

df_botanical_garden["ADDITIONAL DETAILS"] = 'Not Available'
df_botanical_garden
```

```python
#Rearranging the columns in the dataframe

df_botanical_garden = df_botanical_garden[["CATEGORY", "NAME", "ADDRESS", "ZIPCODE", "PHONE NO.", "URL", "ADDITIONAL DETAILS"]]
df_botanical_garden
```

```python
#Checking for missing values

for c in df_botanical_garden.columns:
    miss = df_botanical_garden[c].isnull().sum()
    
    if miss > 0 :
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has NO missing value!".format(c))
```

```python
#Writing the dataframe to a csv file
df_botanical_garden.to_csv('/Users/ramyakumaresan/Downloads/Botanical_Gardens.csv' ,index=False)
```
