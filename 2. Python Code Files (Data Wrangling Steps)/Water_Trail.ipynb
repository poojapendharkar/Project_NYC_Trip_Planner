```python
#Data cleaning for Water_Trail Dataset

#import libraries and dataset
import pandas as pd
from pandas_profiling import ProfileReport

data = pd.DataFrame(pd.read_csv('Water_Trails.csv'))

data
```

```python
#profile the data
profile= ProfileReport(data, title="profile")
profile.to_file("watertrail_report.html")
```

```python
#drop the columns not required for our analysis

data = data.drop(columns=['Belongs_to_Parks', 'Available_Parking', 'Launch___Land'])

data.head()
```

```python
#fill nan values for Description and storage with Not Available

data['Description'] = data['Description'].fillna("Not Available")


data['Hours'] = data['Hours'].fillna("Not Available")

data
```

```python
#concatenate columns to form one Additional Details column 

data['ADDITIONAL DETAILS'] = "Description: " + data['Description'] + " Hours: " + data['Hours'] + " Storage: " + data['Storage']

data.head()
```

```python
data = data.drop(columns=['Description','Hours', 'Storage'])

data.head()
```

```python
#rename columns

data.rename(columns = {'Name':'NAME','Location':'ADDRESS'}, inplace = True)
data.head()
```

```python
data['Address_zip'] = data['NAME'] + ", New York" 

data.head()
```

```python
data.dropna(subset = ['ADDRESS', 'Address_zip'])
```

```python
#Extract the address as dictionaries

wt_address = data[['Prop_ID','Address_zip']]
addresses_dict = wt_address.set_index('Prop_ID').T.to_dict('list')

addresses_dict
```

```python
# extract ziocode from address
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="geoapiExercises")
```
