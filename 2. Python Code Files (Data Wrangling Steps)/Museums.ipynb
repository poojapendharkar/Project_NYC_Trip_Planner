```python
# Import the required libraries
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
# Read the CSV file "museums.csv" into the dataframe "df_museums"
df_museums = pd.read_csv("/Users/ramyakumaresan/Downloads/Museums.csv")
pd.set_option('display.max_rows', 1000)
df_museums
```

```python
#profile the data
profile= ProfileReport(df_museums, title="profile")
profile.to_file("museums_report.html")
```

```python
# Drop the column "the_geom" from the dataframe "df_museums"
df_museums.drop("the_geom", axis=1, inplace=True)
df_museums
```

```python
# Concatenate the columns "ADDRESS1" and "ADDRESS2" into one column "ADDRESS" and then drop the columns "ADDRESS1" and "ADDRESS2"
df_museums["ADDRESS"] = df_museums["ADRESS1"]+" "+df_museums["ADDRESS2"].fillna("")
df_museums.drop(["ADRESS1", "ADDRESS2"], axis=1, inplace=True)
df_museums
```

```python
# Add two new columns "CATEGORY" and "ADDITIONAL DETAILS"
df_museums["CATEGORY"] = "Museums"
df_museums["ADDITIONAL DETAILS"] = "Not available"
df_museums
```

```python
# Rearrange the columns
df_museums = df_museums[["CATEGORY", "NAME", "ADDRESS", "ZIP", "TEL", "URL", "ADDITIONAL DETAILS"]]
df_museums
```

```python
# Rename the columns
df_museums.rename(columns={"ZIP":"ZIPCODE", "TEL":"PHONE NO."}, inplace=True)
df_museums
```

```python
#checking for Null values

df_museums.isnull()
```

```python
#checking for missing values
for c in df_museums.columns:
    miss = df_museums[c].isnull().sum()
    
    if miss > 0 :
        print("{} has {} missing value(s)".format(c,miss))
    else:
        print("{} has NO missing value!".format(c))
```

```python
#Handling missing values for PHONE NO.
df_museums["PHONE NO."] = df_museums["PHONE NO."].fillna("Not available")
df_museums["PHONE NO."]
```

```python
#checking and changing the datatype

df_museums.dtypes
```

```python
#Writing the data into a csv file

df_museums.to_csv('/Users/ramyakumaresan/Downloads/Ramya_Kumaresan_cleaned_museums_dataset.csv', index=False)
```
